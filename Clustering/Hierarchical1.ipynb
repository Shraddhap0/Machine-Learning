```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
data=pd.read_csv("C:\\Users\\Dell\\Desktop\\Data Science\\Assignment 7\\crime_data (1).csv")
data.head()
```

```python
data.drop(["Unnamed: 0"],inplace=True,axis=1)
```

```python
plt.figure(figsize = (20,15));
dendrogram = sch.dendrogram(sch.linkage(data, method='average'))
```

```python
hc = AgglomerativeClustering(n_clusters=6, affinity = 'euclidean', linkage ='average')
hc                            
```

```python
# save clusters for chart
y_hc = hc.fit_predict(data)
Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
Clusters
```

```python
data1 = pd.concat([data,Clusters], axis=1)
data1
```
