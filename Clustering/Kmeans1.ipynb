```python
km=KMeans(n_clusters=12,random_state=25)
```

```python
km.inertia_
```

```python
wcss=[]


for i in range(1,8):
    km=KMeans(n_clusters=i,random_state=42)
    km.fit(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,8),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#Build Cluster algorithm

from sklearn.cluster import KMeans
n_clusters=KMeans(5,random_state=25)
n_clusters.fit(df)
```

```python
n_clusters.labels_
```

```python
df_1=pd.DataFrame(n_clusters.labels_,columns=["cluster_ID"])
df_1
```

```python
Df=pd.concat([df,df_1],axis=1)
Df
```

```python
Df.groupby("cluster_ID").agg(["mean"]).reset_index()
```
