```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data=pd.read_csv("C:\\Users\\Dell\\Desktop\\Data Science\\Assignment 9\\book (2).csv")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

Apriori Algorithm
1. Support = 0.05%, Confidence = 90%

```python
item = apriori(data, min_support=0.05, use_colnames=True)
item
```

```python
rules = association_rules(item, metric="lift", min_threshold=0.8)
rules
```

```python
rules.sort_values('lift',ascending = False)
```

```python
rules[rules.lift>1]
# lift ration may > 1 is good for influential for a rules
```

```python
# visualization of 'Suppoet and confidence' with scatter plot 
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence') 
plt.show()
```

association rule as 5% support and 90% confidence
